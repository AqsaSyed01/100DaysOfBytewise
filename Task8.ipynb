```python
#Task8
#Exercise 1
#importing libraries
from sklearn.datasets import load_iris
import numpy as np
from scipy import stats
import pandas as pd

# Loading the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Extracting the sepal lengths
sepal_lengths = iris_df['sepal length (cm)']

# ean
mean_sepal_length = np.mean(sepal_lengths)

# Median
median_sepal_length = np.median(sepal_lengths)

# Mode
mode_sepal_length_result = stats.mode(sepal_lengths, keepdims=True)
mode_sepal_length = mode_sepal_length_result.mode[0]

mean_sepal_length, median_sepal_length, mode_sepal_length

```

```python
#Exercise2
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd

# Loading the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Extracting the petal widths
petal_widths = iris_df['petal width (cm)']

# Calculate variance
variance_petal_width = np.var(petal_widths, ddof=1)  # ddof=1 for sample variance

# Calculate standard deviation
std_dev_petal_width = np.std(petal_widths, ddof=1)  # ddof=1 for sample standard deviation

variance_petal_width, std_dev_petal_width
```

```python
#Exercise3
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris

# Loading the Iris dataset
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Statistics
summary_table = pd.DataFrame({
    'Mean': iris_df.mean(),
    'Median': iris_df.median(),
    'Variance': iris_df.var(ddof=1),  # Sample variance
    'Standard Deviation': iris_df.std(ddof=1)  # Sample standard deviation
})

# Summary table
print(summary_table)
```

```python
#Exercise4
import pandas as pd
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Extracting the sepal lengths
sepal_lengths = iris_df['sepal length (cm)']

# Frequency distribution
frequency_distribution = sepal_lengths.value_counts().sort_index()

# Probability distribution
total_samples = len(sepal_lengths)
probability_distribution = frequency_distribution / total_samples

# Displaying the probability distribution
probability_distribution = probability_distribution.sort_index()
print(probability_distribution)
```

```python
#Exercise5
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

sepal_lengths = iris_df['sepal length (cm)']

frequency_distribution = sepal_lengths.value_counts().sort_index()


total_samples = len(sepal_lengths)
probability_distribution = frequency_distribution / total_samples

# Plotting Histogram
plt.figure(figsize=(10, 6))
plt.bar(probability_distribution.index, probability_distribution.values, width=0.1, alpha=0.7,
        color='b', edgecolor='black')

plt.title('Probability Distribution of Sepal Lengths in Iris Dataset')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Probability')
plt.grid(True)
plt.show()

```

```python
#Exercise 6
import numpy as np

petal_lengths = iris_df['petal length (cm)']

# Cumulative distribution function (CDF)
sorted_petal_lengths = np.sort(petal_lengths)
cumulative_prob = np.arange(len(sorted_petal_lengths)) / float(len(sorted_petal_lengths))

# Plotting the CDF
plt.figure(figsize=(10, 6))
plt.plot(sorted_petal_lengths, cumulative_prob, marker='o', linestyle='-', color='b')

plt.title('Cumulative Distribution Function (CDF) of Petal Lengths in Iris Dataset')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()

```

```python
#Exercise 8
# Extracting petal lengths
petal_lengths = iris_df['petal length (cm)']

# Threshold value
x = 5.8 

# Calculate the probability
probability_greater_than_x = np.sum(petal_lengths > x) / len(petal_lengths)

print(f"The probability of a randomly selected iris flower having a petal length greater than {x} cm is: {probability_greater_than_x:.2f}")
```

```python
#Exercise 9
from scipy import stats

# Petal lengths for two species (e.g., setosa and versicolor)
setosa_petal_lengths = iris_df.loc[iris.target == 0, 'petal length (cm)']
versicolor_petal_lengths = iris_df.loc[iris.target == 1, 'petal length (cm)']

# Independent samples t-test
t_statistic, p_value = stats.ttest_ind(setosa_petal_lengths, versicolor_petal_lengths)

# Defining significance level
alpha = 0.05

# Results
print(f"T-Statistic: {t_statistic:.4f}")
print(f"P-Value: {p_value:.4f}")

if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in the mean petal length between the two species.")
else:
    print("Fail to reject the null hypothesis: There is no significant difference in the mean petal length between the two species.")
```

```python
#Exercise 10
# Extracting sepal length and sepal width
sepal_length = iris_df['sepal length (cm)']
sepal_width = iris_df['sepal width (cm)']

# Covariance
covariance = np.cov(sepal_length, sepal_width)[0, 1]

# Correlation
correlation = np.corrcoef(sepal_length, sepal_width)[0, 1]

print(f"Covariance between sepal length and sepal width: {covariance:.4f}")
print(f"Correlation between sepal length and sepal width: {correlation:.4f}")
```
